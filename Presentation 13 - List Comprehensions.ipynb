# List Comprehension


---> It provides us a simple way of creating lists using iterable object.(elements could be conditionally included )

---> Contains : 
                1) Output Expression
                
                2) Iterator variable
                
                3) Iterable
                
                
---> With the help of  mathematical notations we implement these.

---> <b> Syntax </b>
 

<b> Examples </b>

```python
side=[1,3,6,9]
area=[i**2 for i in side]
print(area)
```

```python
marks=[87,60,94,32,98,26,30,80,49,63,74]
failed=[i for i in marks if i<36]
print(failed)
```

```python
subjects=["Python4ML", "Java4WebApp", "Python4Web","C4EmbSyst","R4Statistics","SQL4DataBase", "Python4Gaming","ObjC4ios", "Ruby4Scripting","MATLAB4MathReasearch"]
Py=[i for i in subjects if "Python" in i]
print(Py)
```

```python
num=[i for i in range(101)]
print(num)
print(" ")

Multiples_6_below_100 =[i for i in num if i%2==0 if i%3==0]
print(Multiples_6_below_100 )
```

```python
units_consumed = [45,78,23,45,78,12,34,10,7,9,15,17,19]

Electricity_Bills=[j*10 if j <21 else j*15 for j in units_consumed]
print(Electricity_Bills)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 18, 20]
squares = ["small" if number < 10 else "big" for number in numbers]
print(squares)
```

<b> Nested List Comprehension </b>

```python
my_matrix = [[column for column in range(5)] for row in range(3)]
print("Your 3x5 matrix is: ")
print(my_matrix)
```
