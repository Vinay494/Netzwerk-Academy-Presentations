# Scope in Python

---> The part of program till where the variable is accessable.


---> 4 types :
        Local
        
        Enclosing
        
        Global
        
        Build-in  
        
        
        
---> LEGB Rule.
        
    

<b> Local Scope </b>

```python
def loc_scope():
    var="I am Pakka Local"
    print(var)
    
loc_scope()

print(var)
```

<b> Enclosing Scope </b>

```python
def outer():
    first_num = 1
    def inner():
        second_num = 2
        # Print statement 1 - Scope: Inner
        print("first_num from outer: ", first_num)
        # Print statement 2 - Scope: Inner
        print("second_num from inner: ", second_num)
    inner()
    # Print statement 3 - Scope: Outer
    print("----------------------------")
    print("first_num from outer: ", first_num)
    print("second_num from inner: ", second_num)

outer()
```

<b> Global Scope</b>



---> Usually are defined outside the functions

---> Easier to understand

```python
Adjective="Jai"

def Our_Soldier():
    word1="Jawan"
    print(Adjective, word1)
    
def Our_Farmer(word2):
    
    print(Adjective, word2)
    
Our_Soldier()
print("")
Our_Farmer("Kisan")
```

<b> Build-in Scope </b>


---> Widest scope among all

---> All the pre-defined keywords 

<B> LEGB Rule </b>



<b> global keyword</b>

```python
Adj="Jai"

def change_Adj(new):
    global Adj
    Adj=new
    
def farmer():
    word3="Kisan"
    print(Adj, word3)
    
change_Adj("Save")
farmer()
print(Adj)
```
